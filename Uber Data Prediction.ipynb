### Data Manipulation

```python
#Date/Time is in string format and in complex form to analyse. So change in Pandas Date and Time format  
data["Date/Time"] = data["Date/Time"].map(pd.to_datetime)
```

```python
data.head(10)
```

```python
#In new column "DOM" of dataset getting the Day of the Month  
def get_dom(dt):
    return dt.day
data['DOM'] = data["Date/Time"].map(get_dom)
```

```python
data.iloc[12223]
```

```python
#In new column "Weekday" of dataset getting the Weekday 
def get_weekday(dt):
    return dt.weekday()
data["Weekday"] = data["Date/Time"].map(get_weekday)
```

```python
data.tail(2225)
```

```python
#In new column "Hour" of dataset getting the Hour 
def get_hour(dt):
    return dt.hour
data["Hour"] = data["Date/Time"].map(get_hour)
```
